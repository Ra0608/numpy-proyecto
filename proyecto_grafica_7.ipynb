```python
import matplotlib.pyplot as plt
import pandas as  pd
# extraer los datos en un df
df =  pd.read_csv("resultado.csv")
df = df.loc[(df!= '...').any(axis=1)]
df

# Configuración de estilo de seaborn
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Supongo que tienes un DataFrame llamado 'df' con las columnas necesarias

# Cantidad de anémicos
hombres_anemicos = df[(df['anaemia'] == 1) & (df['sex'] == 1)]['age'].count()
mujeres_anemicas = df[(df['anaemia'] == 1) & (df['sex'] == 0)]['age'].count()

# Cantidad de diabéticos
hombres_diabeticos = df[(df['diabetes'] == 1) & (df['sex'] == 1)]['age'].count()
mujeres_diabeticas = df[(df['diabetes'] == 1) & (df['sex'] == 0)]['age'].count()

# Cantidad de fumadores
hombres_fumadores = df[(df['smoking'] == 1) & (df['sex'] == 1)]['age'].count()
mujeres_fumadoras = df[(df['smoking'] == 1) & (df['sex'] == 0)]['age'].count()

# Cantidad de muertos
hombres_muertos = df[(df['DEATH_EVENT'] == 1) & (df['sex'] == 1)]['age'].count()
mujeres_muertas = df[(df['DEATH_EVENT'] == 1) & (df['sex'] == 0)]['age'].count()

# Crear listas para las barras agrupadas
anemicos = [hombres_anemicos, mujeres_anemicas]
diabeticos = [hombres_diabeticos, mujeres_diabeticas]
fumadores = [hombres_fumadores, mujeres_fumadoras]
muertos = [hombres_muertos, mujeres_muertas]

# Configuración de la figura
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar barras agrupadas
bar_width = 0.2
index = np.arange(2)
opacity = 0.8

rects1 = plt.bar(index, anemicos, bar_width, alpha=opacity, color='b', label='Anémicos')
rects2 = plt.bar(index + bar_width, diabeticos, bar_width, alpha=opacity, color='g', label='Diabéticos')
rects3 = plt.bar(index + 2 * bar_width, fumadores, bar_width, alpha=opacity, color='r', label='Fumadores')
rects4 = plt.bar(index + 3 * bar_width, muertos, bar_width, alpha=opacity, color='y', label='Muertos')

# Configuración de la leyenda, ejes y etiquetas
plt.xlabel('CATEGORIAS')
plt.ylabel('Cantidad')
plt.title('Histograma Agrupado por Sexo')
plt.xticks(index + 1.5 * bar_width, )
plt.legend()

# Mostrar la gráfica
plt.tight_layout()
plt.show()

```
