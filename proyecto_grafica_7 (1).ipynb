```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import  plotly.graph_objects as  go

df = pd.read_csv("resultado.csv")

# Cantidad de anÃ©micos
hombres_anemicos = df[(df['anaemia'] == 1) & (df['sex'] == 1)]['age'].count()
mujeres_anemicas = df[(df['anaemia'] == 1) & (df['sex'] == 0)]['age'].count()

# Cantidad de diabÃ©ticos
hombres_diabeticos = df[(df['diabetes'] == 1) & (df['sex'] == 1)]['age'].count()
mujeres_diabeticas = df[(df['diabetes'] == 1) & (df['sex'] == 0)]['age'].count()

# Cantidad de fumadores
hombres_fumadores = df[(df['smoking'] == 1) & (df['sex'] == 1)]['age'].count()
mujeres_fumadoras = df[(df['smoking'] == 1) & (df['sex'] == 0)]['age'].count()

# Cantidad de muertos
hombres_muertos = df[(df['DEATH_EVENT'] == 1) & (df['sex'] == 1)]['age'].count()
mujeres_muertas = df[(df['DEATH_EVENT'] == 1) & (df['sex'] == 0)]['age'].count()

# Crear listas para las barras agrupadas
anemicos = [hombres_anemicos, mujeres_anemicas]
diabeticos = [hombres_diabeticos, mujeres_diabeticas]
fumadores = [hombres_fumadores, mujeres_fumadoras]
muertos = [hombres_muertos, mujeres_muertas]

x = ["AnÃ©micos", "DiabÃ©ticos", "Fumadores", "Muertos"]
y_hombres = [hombres_anemicos, hombres_diabeticos, hombres_fumadores, hombres_muertos]
y_mujeres = [mujeres_anemicas, mujeres_diabeticas, mujeres_fumadoras, mujeres_muertas]

fig = go.Figure()

fig.add_trace(go.Bar(
    x=x,
    y=y_hombres,
    name='Hombres',
    marker_color='blue'
))

fig.add_trace(go.Bar(
    x=x,
    y=y_mujeres,
    name='Mujeres',
    marker_color='red'
))

fig.update_layout(
    barmode='group',
    title='Histograma Agrupado por Categoria y Sexo',
    xaxis_title='CategorÃ­as',
    yaxis_title='Cantidad'
)

fig.show()

```
